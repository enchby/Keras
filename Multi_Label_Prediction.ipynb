##kerasの回帰で複数のラベルを予測

```
from tensorflow.keras import models
from tensorflow.keras import layers
import numpy as np

#６つの説明変数をもつサンプルデータ３セットを用意
train_data = np.array([[17.9,  0.0,  3.4,  0.0,  85,  1017.5],
                                        [20.6,  0.0,  8.2,  0.9,  63,  1017.2],
                                        [21.6,  0.0,  6.0,  0.8,  61,  1015.7]])


#３セットのデータに対応するラベルを用意
#ラベルは３つの数字で構成されている
train_targets =np.array([[4,2,5],　　[6,4,9],　　[8,2,11]])

#モデル定義
model = models.Sequential()
model.add(layers.Dense(64, activation='relu',
                       input_shape=(train_data.shape[1],)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
#最後のDenseをラベルの目的変数の数と合わせる
model.add(layers.Dense(3))
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

#モデルのトレーニング
model.fit(train_data, train_targets,epochs=500)
```

```
#ラベルの目的変数が複数の場合はreshape(1,-1)する必要はない
#予測用データの用意
pre = np.array([[21.0,  0.0,  5.8,  0.4,  62,  1015.2]])

#予測
print (model.predict(pre))
```

##適当なデータセットだが、3つの数値が大きなズレなく予測されている
